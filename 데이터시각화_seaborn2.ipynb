데이터 분포 살펴보기(수치형 Vs 카테고리형)

boxplot( data = df, x = , y = , hue = )

```python
import seaborn as sns

raw = sns.load_dataset('tips')
raw.head(5)
```

```python
# boxplot을 이용해 가로축은 day 컬럼, 세로축은 tip값의 분포로 그릴수 있다.
# 특정 기준에 따른 데이터 분포를 살펴보기 유용하다.
# 데이터 분포는 표시하지만 해당하는 데이터 갯수를 표현하진 않아서, 카테고리마다 데이터 갯수가 상이할경우
# 그대로 해석해서 받아들이긴 어렵다.
sns.boxplot(data = raw, x = 'day', y = 'tip')

# 박스 가운데 선이 중위수임. 위쪽이나 아래쪽 점들은 일반적이지 않은 값들
```

```python
# hue 옵션을 추가하여, 특정 기준의 변화에 따른 데이터 분포를 살펴봄

sns.boxenplot(data = raw, x = 'day', y = 'tip', hue = 'smoker')
```

```python
#boxplot의 단점(데이터 분포는 표현 but 데이터 갯수는 표현x)을 보완하기위해
#swarmplot을 이용한다.

sns.swarmplot(data = raw, x = 'day', y = 'tip')
```

```python
sns.swarmplot(data = raw, x = 'day', y = 'tip', hue = 'smoker')
```

```python
sns.swarmplot(data = raw, x = 'day', y = 'tip', hue = 'smoker', dodge = True)
```

```python
#boxplot 과 swarmplot을 함께 그려볼 수 있다.
sns.boxplot(data = raw, x = 'day', y = 'tip', hue = 'smoker',)
sns.swarmplot(data = raw, x = 'day', y = 'tip', hue = 'smoker', dodge = True)
```

```python
# size / 성별 별 tip의 분포를 살펴보기

sns.boxplot(data = raw, x = 'size', y = 'tip', hue = 'sex')
```

```python
# barplot의 경우 특정값에 대한 개수를 살펴 볼수 있다.

sns.barplot(data = raw, x = 'size', y = 'tip')
```

```python
sns.barplot(data = raw, x = 'size', y = 'tip', hue = 'sex')
```

heatmap( data = df )

```python
raw.head()
```

```python
# 히트맵을 이용하면, 두 카테고리형 데이터 분포에 대한 수치형 데이터의 값을 색상이용 한눈에 알아보기 쉽다.
# 요일별, size별 평균 tip 데이터를 가진 피벗테이블을 먼저 만들어준다.

df = raw.pivot_table(index = 'day', columns = 'size', values = 'tip', aggfunc = 'mean')
df
```

```python
# 보고자하는 관점에 따라 데이터 프레임을 만든 후, heatmap()함수를 이용하면 히트맵을 작성 할 수 있다,

sns.heatmap(data = df,)
```

```python
# 실제 데이터를 확인하기 위해 히트맵 내에 수치를 표현 할 수도 있다.
# annot = true로 할 경우 수치를 표현하며, 표현되는 숫자 형태는 fmt옵션으로 지정 가능
# fmt = '2f'의 경우 0 두 소수점 두자리까지 표현하라는 의미이며,
# ".0f"의 경우 소수점 뒤에 표현되는 자리수 없이 정수로만 표현하라는 의미.
# fmt = 포맷팅 .을 쓰고 표현하고 싶은 소수점 자리를 숫자로 쓴다.
sns.heatmap(df, annot = True,)
```

```python
sns.heatmap(df, annot = True,  fmt = '.3f')
```

```python
sns.heatmap(df, annot = True,  fmt = '.0f')
```

```python
#cmap(컬러맵) 옵션을 이용해, 색상을 조정할 수도 있다
#seaborn에서는 다양한 종류의 컬러차트를 지원하며 이중 blues를 이용해보겠다
# 추천색상은 : reds, blues, vlag, pastel1 RdBu_r)
sns.heatmap(data = df, annot = True, fmt = '.0f', cmap = 'Blues')
```

```python
sns.heatmap(data = df, annot = True, fmt = '.0f', cmap = 'RdBu_r')
# _r은 리버스 컬러별 나타내는거 바꿀 수 있음.
```
