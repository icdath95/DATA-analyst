```python
# 데이터 시각화를 위해 seabron 라이브러리 불러오기

import seaborn as sns
```

```python
# seaborn에는 예제로 활용할 수 있는 데이터 셋이 연결되어있음. tips 데이터셋 불러오기
# 불러온 데이터 셋은 판다스 데이터프레임 형식으로 판다스와 동일한 명령어 사용가능

raw = sns.load_dataset('tips')
```

```python
# head를 통해 데이터 살펴보기

raw.head()
```

```python
# info를 통해 데이터 구성 / 형식 살펴보기

raw.info()
```

####데이터 분포 살펴보기(수치형 vs 수치형)

### `relplot`( data = df, x = ,  y = , hue = , kind = 'scatter')
두 개의 변수(모두 수치형 데이터) 분포를 확인할 때

- kind 옵션
    - 'scatter' (기본값)
    - 'line'

```python
# sns.relplot() 을 이용해 두 데이터간의 "관계"를 시각화 할 수 있다.
# kind 옵션을 이용해 표현방식을 조정할 수 있는데, 기본 설정값은 'scatter'이다.

sns.relplot(data = raw, x = 'tip', y = 'total_bill')
```

```python
# kind 옵션을 line 으로 조정할 경우 선그래프 가능하다.

sns.relplot(data = raw, x = 'tip', y = 'total_bill', kind = 'line')
```

```python
# 성별에 따라 색상 구분하기
# hue 옵션을 이용해 색상의 차이로 종류를 구분할 수 있다.

sns.relplot(data = raw, x = 'tip', y = 'total_bill', hue = 'sex')
```

###jointplot( data = df, x = , y = , kind = 'scatter')

kind에 따라 그래프 형대 변경됨

"scatter"(기본값): point

"reg": point + regression

"kde": 누적 분포 차트 like 지도

```python
#jointplot을 이용해 두 수치데이터간의 관계와, 각 데이터 분포를 함께 확인 가능
#kind 옵션을 통해 여러 타입으로 시각화 할 수 있는데, 기본적으로 kind = 'scatter' 로 지정

sns.jointplot(data = raw, x = 'tip', y = 'total_bill')
```

```python
# kind = 'kde'로 지정할 경우, 밀도함수를 이용해 시각화 가능
# 진하게 표시될 수록 집중되어있다는것을 의미 값이 급격하게 변할 수록 선의 간격이 좁아짐

sns.jointplot(data = raw, x = 'tip', y = 'total_bill', kind = 'kde')
```

```python
# kind = 'regg' 옵션을 사용할 경우 회귀선을 확인할 수 있다.

sns.jointplot(data = raw, x = 'tip', y = 'total_bill', kind = 'reg')
```

```python
#kind = hex 옵션은 kde 속성과 비슷하게 밀도를 표현하나 kde처럼 간격이 다르지 않고
# 모두 동일한 간격의 육각형으로 표현된다.

sns.jointplot(data = raw, x = 'tip', y = 'total_bill', kind = 'hex')
```

### `pairplot`( data = df)
df의 모든 수치형데이터 컬럼에서 "두 컬럼"씩 관계를 시각화 합니다. 

```python
# pairplot을 이용해 total_bill, tip, size 데이터 간의 관계를 알아봄.

sns.pairplot(data = raw)
```

```python
# pairplot에서도 hue 를 이용해 특정 컬럼의 값에 따른 분포를 색상을 이용해 시각화 할 수 있다.

sns.pairplot(data = raw, hue = 'sex')
```
