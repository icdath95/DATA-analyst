```python
# API를 이용해 자료를 받아오기 위해 requests 라이브러리, 자료저장을 위해 pandas 불러오기

import requests 
import pandas as pd
```

```python
# 서울열린데이터광장에서 받은 API 인증키를 입력.

apikey = "56584d4148707070313034564251536e"
```

```python
# 요청할 URL 주소를 만든다.
# 여기선 한번에 1000개만 요청 가능함, 2번에 나눠서 가져옴
# 주소를 2개써도 되지만 여기선 f로 해본다.
startnum = 1
endnum = 1000
# 
url = f'http://openapi.seoul.go.kr:8088/{apikey}/json/bikeList/{startnum}/{endnum}/'
print(url)
```

```python
# 자료 요청시
json1 = requests.get(url).json()

# response[200] = > 이상없다는 뜻.
# 크롤링 = .text / 음성, 영상 = .content
```

```python
json1 
```

```python
# 필요한 정보를 선택
json1.keys()
```

```python
# 필요한 정보를 선택
json1['rentBikeStatus'].keys()
```

```python
# 데이터 구조를 확인
# 데이터의 개수

json1['rentBikeStatus']['list_total_count']
```

```python
# 데이터 구조를 확인
# 오류 여부 화깅ㄴ
json1['rentBikeStatus']['RESULT']
```

```python
#데이터 구조를 확인
# 자전거 정류장별 자전거 현황
json1['rentBikeStatus']['row']
```

```python
# 판다스의 데이터 프레임으로 전환

raw1 = pd.DataFrame(json1['rentBikeStatus']['row'])
raw1.head()
```

```python
# 다음 1001부터 2000까지 데이터 가져오기
startnum = 1001
endnum = 2000

url2 = f'http://openapi.seoul.go.kr:8088/{apikey}/json/bikeList/{startnum}/{endnum}/'

json2 = requests.get(url2).json()
```

```python
json2
```

```python
raw2 = pd.DataFrame(json2['rentBikeStatus']['row'])
raw2.head()
```

```python
# 두 데이터 통합
# 지금은 위아래로 이어서 붙여야 하기때문에 merge말고 append 사용

data = raw1.append(raw2)
data.head()
```

```python
data.info()
```

```python
# 엑셀로 저장
# index 저장 안하려면 밑에처럼.
data.to_excel("./bycicle.xlsx",index = False)
```
