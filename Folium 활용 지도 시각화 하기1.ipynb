## 지도 시각화
- 지도생성하기
    - `m` = folium.Map(location = [위도, 경도],  zoom_start = 확대정도)

- 정보 추가하기
    - 마커 추가하기
        - folium.Marker([위도, 경도]).add_to(`m`)
    - 원 추가하기
        -     folium.CircleMarker([위도, 경도],radius= 원크기).add_to(`m`)
    - 추가옵션: 
        - tooltip="마우스 올리면 보여질 정보"
        - popup="클릭하면 보여질 정보"          
    - 기타) ClickForMarker('체크').add_to(`m`)  지도에서 클릭할 경우 마커 추가하기

```python
## 1. 지도생성하기.
# 먼저, 지도를 펼치기  지도에서 어느 부분을 살펴볼 건지, 지도 중심의 좌표를 위도, 경도를 이용해 지정 
# 서울역을 기준으로 살펴봄 [37.5536067,126.9674308]
# zoom_start 옵션을 통해, 처음에 얼마나 확대해서 볼 것인지를 지정 
# 지도가 생성된 뒤에 위치 이동이나 확대/축소를 조정 가능

m = folium.Map(location=[37.5536067,126.9674308], zoom_start=12)
m
```

```python
# 2.마커추가하기

# Marker( [위도, 경도]).add_to(지도이름) 명령을 통해 지도상에 마커를 추가할 수 있으며
# tooltip, popup 옵션을 통해 설명을 추가할 수 있습니다.

folium.Marker([37.5536067,126,9674308],
              tooltip="마우스",# 마우스올리면 정보올라감
              popup="클릭").add_to(m)#클릭하면 정보뜸
m
```

```python
#circlMaker([위도,경도]).add_to(지도이름) 명령을 통해, 동그라미를 지도에 표시
#이때 radius 옵션을 통해 동그라미의 크기를 지정 할 수 있다.
# 서울역: [37.5536067,126.9674308]

# 써클마커 추가하기 = 마켜와 똑같이 tooltip, popup 모두 사용가능
folium.CircleMarker([37.5536067,126.9674308],radius=20).add_to(m)
m
```

## 미니맵을 추가할 경우 MiniMap

```python
#folium에는 다양한 효과를 줄 수 있는 plugings가 존재하는데, minimap도 그중 하나
# 미니맵을 지도 우측 하단 추가하여 현재 어느위치를 살펴보고 있는지 점검 가능
# 서울역: [37.5536067,126.9674308]

from folium.plugins import MiniMap

m = folium.Map(location = [37.5536067,126.9674308], zoom_start = 12)

# 미니맵 추가하기
minimap = MiniMap()  # MM 다 대문자
minimap.add_to(m)

m
```

```python
# 모든 정보를 처음부터 지도에 추가하지 않고, 이후에 마우스를 이용해 추가할 수 있도록 만들 수도 있습니다. 

# 지도 생성하기
m = folium.Map(location=[37.5536067,126.9674308],   # 기준좌표: 서울역
               zoom_start=12)

# 써클마커 추가하기
folium.CircleMarker([37.5536067,126.9674308],
                    radius=20,
                    tooltip = '마우스올릴경우'
               ).add_to(m)
folium.ClickForMarker('체크추가').add_to(m)    
m
```

## 서울 대피소 현황 지도 만들기

서울 대피소 현황자료를 다운받아 이를 지도에 시각화 해보겠다.

```python
# 먼저 csv 파일의 표 형태 데이터를 읽어오기 위해 판다스 불러오기.

import pandas as pd
```

```python
# 서울열린데이터광장에 있는 서울시 대피소 현황자료 다운 후 판다스로 조회
# 서울 대피소 현황자료 다운받기 = 난 없어서 서울시 지진대피소 현황으로
# http://data.seoul.go.kr/dataList/OA-2189/S/1/datasetView.do
file = '/content/서울시 대피소 방재시설 현황 (좌표계_ WGS1984).csv'
raw = pd.read_csv(file, encoding = 'cp949')   # encoding = 'cp949' : MS 프로그램 사용시 / 그외의 경우 encoding = 'utf-8'  (기본값)
```

## 마커가 너무 많으면 살펴보기 힘듬
# ClusterMarker를 이용해 근처에 있는 마커들끼리 그룹으로 표현

```python
#MakerCluster 라이브러리를 불러오겠습니다.

from folium.plugins import MarkerCluster
```

```python
# MarkerCluster 를 이용해 대피소 정보를 지도에 시각화 하겠습니다. 

# 지도 생성하기
m = folium.Map(location=[37.5536067,126.9674308],
               zoom_start=12)
marker_cluster = MarkerCluster().add_to(m)  # 클러스터 추가하기

# 대피소 마커 추가하기

for i in range(len(raw)):
    lat = raw.loc[i, '위도']
    long = raw.loc[i, '경도']
    name = raw.loc[i,'대피소명칭']
    folium.Marker([lat, long],
                tooltip = name).add_to(marker_cluster)    
m
```

## 미니맵 만들기

```python
# 미니맵 추가하기
from folium.plugins import MiniMap
```

```python
# 미니맵과 클러스터 마커를 이용하여 서울대피소 지도 만들기 

# 위치명 / 수용인원 등 추가 정보 추가하여 지도 만들기
# 지도 생성하기
m = folium.Map(location=[37.5536067,126.9674308],
               zoom_start=12)
# 클러스터 만들기
marker_cluster = MarkerCluster().add_to(m)

# 미니맵 추가하기 #
minimap = MiniMap() 
m.add_child(minimap)

# 대피소 마커 추가하기

for i in range(len(raw)):
    lat = raw.loc[i, '위도']
    long = raw.loc[i, '경도']
    name = raw.loc[i,'대피소명칭']
    folium.Marker([lat, long], tooltip = name).add_to(marker_cluster)    
m
```
