### 3) 아래의 특성들을 만드세요

1. **TotalMntProducts** - 최근 2년간 구매한 모든 제품 가격들을 합한 값을 가진 계산한 특성입니다

2. **TotalNumPurchases** - 구매횟수를 모두 합친 값을 가진 특성입니다

3. **DaysSinceJoined** - 오늘 날짜를 기준으로 가입한 기간을 보여주는 특성입니다 

4. **VIP** - 누적 구매 금액이 $2,000 이상을 넘고, 오늘을 기준으로 가입기간이 5년 이상이 된다면 1, 아니면 0의 값을 가진 칼럼을 만듭니다

```python
data = data.drop("Unnamed: 0", axis = 1)
data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer'])
data.head()
```

```python
# 답안 작성은 이 곳에 하십시오
data["TotalMntProducts"] = data.iloc[:, 10:16].sum(axis = 1)
data["TotalNumPurchases"] = data.iloc[:, 16:19].sum(axis = 1)
data["DaysSinceJoined"] = pd.datetime.now()-data["Dt_Customer"]

data.head()
```

```python
data["DaysSinceJoined"]= pd.to_numeric(data["DaysSinceJoined"].dt.days, downcast='integer')
data.info()
```

```python
data.head()
```

```python
#VIP - 누적 구매 금액이 $2,000 이상을 넘고, 오늘을 기준으로 가입기간이 5년 이상이 된다면 1, 아니면 0의 값을 가진 칼럼을 만듭니다
data.assign(VIP = "" )
conditions = [(data["TotalMntProducts"] >= 2000) & (data["DaysSinceJoined"] >= 1825),
              (data["TotalMntProducts"] < 2000) & (data["DaysSinceJoined"] < 1825)]

# create a list of the values we want to assign for each condition
values = ["1", "0"]

# create a new column and use np.select to assign values to it using our lists as arguments
data["VIP"] = np.select(conditions, values)

data.head()
#data["VIP"] = data["TotalMntProducts"] > 2000 and data["DaysSinceJoined"] > 1825
#(df['age'] > 40) & (df['sex'] == 'Male')
```
