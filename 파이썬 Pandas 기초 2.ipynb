남자 / 여자 비율 차이가 적을수록 --> 성별구분이 없는 이름이지 않을까?

---



```python
# 남자/여자 이름 등록수 합계를 계산하고
## 'sum' 칼럼 추가하기

name_df['M'] + name_df['F']
```

```python
name_df['Sum'] = name_df['M'] + name_df['F']
name_df.head()
```

```python
# 남성/여자 등록비율을 계산한다.

name_df['F_ratio'] = name_df['F'] / name_df['Sum']
name_df['M_ratio'] = name_df['M'] / name_df['Sum']

name_df.head()
```

```python
# 두 비율의 차이를 계산해서 차이가 적을수록 남/녀 구분없이 사용되는 이름인걸 추론
## abs() => 절대값을 보여줌! -+ 상관없이
### 그래서 데이터프레임에 컬럼을 하나더 추가해준다 M_F_GAP

name_df['F_ratio'] - name_df['M_ratio']


```

```python
# 절대값 전환
abs(name_df['F_ratio'] - name_df['M_ratio'])
```

```python
# 컬럼추가
name_df['M_F_GAP'] = abs(name_df['F_ratio'] - name_df['M_ratio'])
name_df.head()
```

```python
# 이름이 가장 많이 사용된 수를 기준으로 내림차순으로 정렬하기
## Sum 기준 내림차순

name_df = name_df.sort_values(by='Sum',ascending=False)
name_df.head()
```

```python
#남자/여자 사용비율의 차이가 0.1보다 작은 경우를 찾는다.
## name_df[m-f-gap] < 0.1

# 내가 원하는 조건 만들기
cond = name_df['M_F_GAP'] < 0.1
# 원하는 조건에 부합하는 데이터만 보여주기
name_df[cond]
```

```python
# 위 조건문에 부합하면서 상위 10개만 보이기
name_df[cond].head(10)
```

세대 기준으로 그룹 만들기
  - 한 세대 나누는 기준 30년: 2020년 기준 30년씩 구분
  - 1930년대 이전
  - 1960년대 이전
  - 1990년대 이전
  - 2020년 이전

```python
# 출생연도 시리즈에서 순서대로 해당하는 세대 그룹명에 매칭하고 그 결과를 리스트에 저장함.
## 1930년 이전일 경우 -> 1930년 이전
## 1960년 이전일 경우 -> 1960년 이전
## 1990년 이전일 경우 -> 1990년 이전
## 그외 - > 2020년 이전

### 4그룹으로 나눠서 저장할 곳은 아래 리스트 이고 []만 넣어서 빈 리스트를 만들어준다
### 이후에 반복문을 사용해서 그룹을 나눠서 값을 넣어준다.
year_class_list = []
for year in raw['YearOfBirth']:
    if year <=1930:
      year_class='1930년이전'
    elif year <=1960:
      year_class = '1960년이전'
    elif year <=1990:
      year_class = '1990년이전'
    else: 
      year_class = '2020년이전'

    year_class_list.append( year_class)
```

```python
year_class_list
```

```python
raw['year_class'] = year_class_list

raw.head()
```

전체 컬럼 합계 계산하기

옵션1) 모든 컬럼을 하나씩 더하기: df['컬럼1'] + df['컬럼2'] + ... + df['컬럼n']

옵션2) sum() 활용하기: df.sum(axis = 1)

참고) df.sum() 을 활용하면, 기본값으로 axis = 0 으로 지정되며, 컬럼별 합계가 아닌 row 별 합계가 계산됨

```python
name_period.head()
# 위 DF의 name/sex에 대한 컬럼별 합계를 구하고 싶을떄
#옵션1)의 방법은 직접 컬럼이름을 하나씩 써서 구한다.

#ex) name_perod['1990년 이전'] + name_perion[1960년 이전'] ...
# => 그러면 aadan의 1990년 이전, 1930년 이전 .... 의 합계가 나온다.
```

```python
#옵션2)의 방법은 sum을 이용하여 컬럼별 합계를 추가한다.
# sum(axis = 1)을 활용하여 컬럼별 합계를 추가합니다. 

name_period.sum()
# => 합계가 나오는데 컬럼의 합계가 나옴.
# 즉 ['1990년 이전']의 모든 데이터가 합쳐져서 ['1990년 이전'] 맨 아랫줄에 데이터가 나옴
# 행의 합계가 나옴 컬럼의 세로 합계
# sum뒤에 아무거도 없거나 axis 아무값 설정안하면 위와 같음
```

```python
#우리가 원하는 값은 열의 합계임. 4개의 컬럼의 합계를 도출해서 가로 맨끝에 두기

# nmae_perion.iloc[ 0, 1 ]
# iloc 복습하기.

name_period.sum(axis= 1)
```
